### 4. Go Forth and Multiply (15 pts)
```
Multiplication table: The multiplication table summarizes the products of varying
combinations of 2 natural numbers. An example is shown below:

      1   2  3    4   5   ...
 1    1   2  3    4   5   ...
 2    2   4  6    8   10  ...
 3    3   6  9    12  15  ...
 4    4   8  12   16  20  ...
 5    5   10 15   20  25  ...
... ... ... ... ...  ...  ...

Requirement: Create a function ‘multtable’ that accepts one required parameter size and
prints out a size x size multiplication table (without the column and row labels).```

````python
size = int(input("Size: "))
h= range(0,size)

def multtable(size):
    i = 0
    while(i < size):

        # nested loop
        j = 0
        while(j < size):
            if i == 0:
                print(h[j], end="  ")
            elif j == 0:
                print(i, end="  ")
            else:
                print(i*j, end="  ")
            j += 1

        print()
        i += 1

multtable(size)
````

```

Do you prefer (a) cats, (b) dogs, or (c) fish?

b. If the user input is not either ‘a,’ ‘b’, or ‘c’ for any of the questions, print “Sorry, you
inputted a wrong letter. Please start again.” then ask each of the questions again,
starting from the first question.

c. For each question, each response corresponds to an element. The possible
equivalents are Fire, Water, and Earth. The corresponding element for each choice

Q1: mountains = fire, city = earth, beach = water
Q2: yoghurt = earth, ice cream = water, gelato = fire
Q3: reading = water, singing = fire, writing = earth
Q4: sweet = earth, salty = water, spicy = fire
Q5: cats = earth, dogs = fire, fish = water

choice. The element with the most number of points by the end of the quiz is the
element of the user.

solve the tie.
If there is a tie between fire and earth, earth would win
If there is a tie between fire and water, fire would win
If there is a tie between earth and water, water would win.```

````python
fire = 0
earth = 0
water = 0

while(True):
    Q1 = input("Do you prefer (a) the mountains, (b) the city, or (c) the beach? ")
    if Q1 == "a" or Q1 == "b" or Q1 == "c":
        if Q1 == "a":
            fire += 1
        elif Q1 == "b":
            earth += 1
        elif Q1 == "c":
            water += 1    
        else:
            print("Sorry, you inputted a wrong letter. Please start again.")
        
        Q2 = input("Do you prefer (a) yoghurt, (b) ice cream, or (c) gelato? ")
        if Q2 == "a" or Q2 == "b" or Q2 == "c":
            if Q2 == "c":
                fire += 1
            elif Q2 == "a":
                earth += 1
            elif Q2 == "b":
                water += 1
            
            Q3 = input("Do you prefer (a) reading, (b) singing, or (c) writing? ")
            if Q3 == "a" or Q3 == "b" or Q3 == "c":
                if Q3 == "b":
                    fire += 1
                elif Q3 == "c":
                    earth += 1
                elif Q3 == "a":
                    water += 1
    
            
                Q4 = input("Do you prefer (a) sweet, (b) salty, or (c) spicy food? ")
                if Q4 == "a" or Q4 == "b" or Q4 == "c":
                    if Q4 == "c":
                        fire += 1
                    elif Q4 == "a":
                        earth += 1
                    elif Q4 == "b":
                        water += 1
                        
            
                    Q5 = input("Do you prefer (a) cats, (b) dogs, or (c) fish? ")
                    if Q5 == "a" or Q5 == "b" or Q5 == "c":
                        if Q5 == "b":
                            fire += 1
                        elif Q5 == "a":
                            earth += 1
                        elif Q5 == "c":
                            water += 1
                        
                        if fire>earth and fire>water:
                            print("Your element is Fire")
                        elif earth>fire and earth>water:
                            print("Your element is Earth")
                        elif water>fire and water>earth:
                            print("Your element is Water")
                        elif fire == earth:
                            print("Your element is Earth")
                        elif fire == water:
                            print("Your element is Fire")
                        elif earth == water:
                            print("Your element is Water")
                        break
                    else:
                        print("Sorry, you inputted a wrong letter. Please start again.")
                else:
                    print("Sorry, you inputted a wrong letter. Please start again.")

            else:
                print("Sorry, you inputted a wrong letter. Please start again.")

        else:
            print("Sorry, you inputted a wrong letter. Please start again.")

    else:
        print("Sorry, you inputted a wrong letter. Please start again.")
````
