```python
#write code here

#phone column
phone = recweek['phonenumber']

#check unique values
phone.unique()

#checks the variety of number of characters phonenumbers have
phone.str.len().value_counts()
```

```python
# write code here

#Provider Codes from Google
GlobeCodes = ['905', '906', '977', '915', '916', '926', '927', '935', '936', '937', '996', '997', '917']
SmartCodes = ['920', '930', '938', '939', '907', '908', '909', '910', '912', '919', '921', '928', '929', '947', '948', '948', '989', '918', '999']

#newdf for gender and provider
carrier = final_recweek[['gender']]
carrier = carrier.copy()
```

```python
#determine if Globe or Smart
carrier['carrier'] = ["Globe" if x[1:4] in GlobeCodes
                      else "Smart" if x[1:4] in SmartCodes
                      else "idk"
                      for x in final_recweek['phonenumber']]
carrier
```
